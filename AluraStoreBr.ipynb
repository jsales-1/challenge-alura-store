### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

lojas = [loja,loja2,loja3,loja4]

loja.head()

```

# 1. Análise do faturamento


```python
def faturamento_loja(alguma_loja):
   valor_faturamento = sum(alguma_loja['Preço'])
   return valor_faturamento
```

```python
for i in range(len(lojas)):
   faturamento = faturamento_loja(lojas[i])
   print(f"Faturamento da loja {i} = {faturamento} reais")
```

# 2. Vendas por Categoria


```python
def categorias_mais_populares_por_loja(alguma_loja):
    mais_populares = dict(alguma_loja['Categoria do Produto'].value_counts())
    return mais_populares
```

```python
for i in range(len(lojas)):
   categorias_e_venda_loja = categorias_mais_populares_por_loja(lojas[i])
   print(f"Vendas por Categoria {i} = {categorias_e_venda_loja}\n")
```

# 3. Média de Avaliação das Lojas

```python
def media_avaliacoes_loja(algum_loja):
   media_avaliacao = sum(algum_loja['Avaliação da compra'])/len(algum_loja)
   return media_avaliacao
```

```python
for i in range(len(lojas)):
   media_avaliacao = media_avaliacoes_loja(lojas[i])
   print(f"Média de Avaliação da Loja {i} = {media_avaliacao}\n")
```

# 4. Produtos Mais e Menos Vendidos

```python
def mais_e_menos_vendidos(alguma_loja):
   produtos = alguma_loja['Produto']
   dict_max_min = {}
   
   dict_max_min["Mais vendido"] = f"{produtos.idxmax()} tendo {produtos.max()}"
   dict_max_min["Menos vendido"] = f"{produtos.idxmin()} tendo {produtos.min()}"

   return dict_max_min
```

```python
for i in range(len(lojas)):
   vendas_max_min = mais_e_menos_vendidos(lojas[i])
   print(f"Produtos mais e menos populares da Loja {i} = {vendas_max_min}\n")
```

# 5. Frete Médio por Loja


```python
def valor_medio_frete(algum_loja):
   frete_medio = sum(algum_loja['Frete'])/len(algum_loja)
   return frete_medio
```

```python
for i in range(len(lojas)):
   media_frete = valor_medio_frete(lojas[i])
   print(f"Valor médio de frete da Loja {i} = {media_frete}\n")
```

# Relatório Final: Recomendação de Venda de Loja

## Introdução
Este relatório tem como objetivo analisar o desempenho de quatro lojas (Loja 0, 1, 2 e 3) para recomendar qual delas o Senhor João deve vender. Foram considerados os seguintes critérios: **faturamento total**, **vendas por categoria**, **média de avaliações**, **produtos mais/menos vendidos** e **frete médio**. A análise busca equilibrar desempenho financeiro, satisfação do cliente e eficiência operacional.

---

## Análise dos Dados

### 1. Faturamento Total
- **Loja 0**: R$ 1.534.509,12  
- **Loja 1**: R$ 1.488.459,06  
- **Loja 2**: R$ 1.464.025,03  
- **Loja 3**: R$ 1.384.497,58  

A **Loja 0** lidera em faturamento, seguida pelas Lojas 1, 2 e 3. Entretanto, a diferença entre a primeira e a última é de aproximadamente R$ 150.000, sugerindo que todas possuem receitas significativas.

---

### 2. Categorias de Produtos Mais Vendidas
- **Loja 0**: Moveis (465 vendas), Eletrônicos (448).  
- **Loja 1**: Moveis (442), Eletrônicos (422).  
- **Loja 2**: Moveis (499), Eletrônicos (451).  
- **Loja 3**: Moveis (480), Eletrônicos (451).  

Todas as lojas têm **móveis** e **eletrônicos** como categorias dominantes. A **Loja 2** destaca-se pela maior venda de móveis, enquanto a **Loja 3** tem menor participação de "eletrodomésticos", possivelmente indicando uma lacuna em diversificação.

---

### 3. Média de Avaliações dos Clientes
- **Loja 0**: 3,98  
- **Loja 1**: 4,04  
- **Loja 2**: 4,05  
- **Loja 3**: 3,99  

As **Lojas 1 e 2** possuem as melhores avaliações, refletindo maior satisfação dos clientes. A **Loja 0** tem a menor média, o que pode impactar a fidelização.

---

### 4. Produtos Mais e Menos Vendidos
- **Mais vendido**: "Xadrez de madeira" em todas as lojas.  
- **Menos vendido**: "Assistente virtual" em todas as lojas.  

O padrão sugere que o **Xadrez de madeira** é um produto estratégico, enquanto "Assistente virtual" pode exigir revisão de estoque ou estratégias de promoção.

---

### 5. Frete Médio por Loja
- **Loja 0**: R$ 34,69  
- **Loja 1**: R$ 33,62  
- **Loja 2**: R$ 33,07  
- **Loja 3**: R$ 31,28  

A **Loja 3** tem o **menor frete médio**, indicando potencial eficiência logística. Já a **Loja 0** possui o custo mais alto, o que pode reduzir competitividade.

---

## Conclusão e Recomendação

### Loja Recomendada para Venda: **Loja 3**

#### Justificativa:
1. **Faturamento mais baixo**: Apesar de ter receita significativa (R$ 1,38 mi), é a menor entre as quatro, sugerindo menor potencial de crescimento comparativo.  
2. **Avaliação mediana**: Com 3,99, está abaixo das Lojas 1 e 2, indicando possíveis desafios na experiência do cliente.  
3. **Diversificação de categorias**: Menor participação em "eletrodomésticos" pode limitar oportunidades em um mercado relevante.  



```python
# Célula adicional para visualizações no Jupyter Notebook

import matplotlib.pyplot as plt

# Configurações gerais
plt.style.use('ggplot')
cores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# 1. Gráfico de Barras: Faturamento por Loja
faturamentos = [faturamento_loja(loja) for loja in lojas]
lojas_labels = [f'Loja {i}' for i in range(4)]

plt.figure(figsize=(10, 5))
plt.bar(lojas_labels, faturamentos, color=cores, edgecolor='black')
plt.title('Faturamento Total por Loja (R$)', fontsize=14, pad=20)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Faturamento (milhões R$)', fontsize=12)
plt.ylim(1_300_000, 1_600_000)
plt.gca().yaxis.set_major_formatter('R${x:,.0f}')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ------------------------------------------------------------------------------------

# 2. Gráfico de Pizza: Distribuição de Categorias na Loja 3 (Recomendada para venda)
categorias_loja3 = categorias_mais_populares_por_loja(lojas[3])

plt.figure(figsize=(8, 8))
plt.pie(
    categorias_loja3.values(),
    labels=categorias_loja3.keys(),
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Paired.colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 0.5}
)
plt.title('Distribuição de Categorias - Loja 3', fontsize=14, pad=20)
plt.axis('equal')
plt.show()

# ------------------------------------------------------------------------------------

# 3. Gráfico de Dispersão: Relação Preço vs. Frete (Todas as Lojas)
dados_completos = pd.concat(lojas)  # Combina dados de todas as lojas

plt.figure(figsize=(10, 6))
for i, loja in enumerate(lojas):
    plt.scatter(
        loja['Preço'], 
        loja['Frete'], 
        color=cores[i], 
        alpha=0.6,
        label=f'Loja {i}'
    )
plt.title('Relação entre Preço do Produto e Custo do Frete', fontsize=14, pad=20)
plt.xlabel('Preço do Produto (R$)', fontsize=12)
plt.ylabel('Custo do Frete (R$)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.gca().xaxis.set_major_formatter('R${x:,.0f}')
plt.gca().yaxis.set_major_formatter('R${x:,.0f}')
plt.show()
```
